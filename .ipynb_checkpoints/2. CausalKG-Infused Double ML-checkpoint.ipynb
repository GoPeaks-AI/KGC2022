In this example, I illustrate how Neo4j-based Causal Knowledge Graph (Causal KG) can work together with the [DoubleML](https://docs.doubleml.org/stable/index.html) package to estimate the effect of 401(k) eligibility and participation on accumulated assets. 

401(k) plans are pension accounts sponsored by employers. The key problem in determining the effect of participation in 401(k) plans on accumulated assets is that many key factors can affect both a family's 401(k) eligibility and their net financial wealth, thus posting a confounding bias. I will show how to debias this confounding from the measures based on a causal KG. I also show that the estimates may be sensitive to the structure of the causal KG. Thus, a well-informed causal KG will help us identify the true causal estimates.

*Define a function for connecting to an expert Causal KG database to query a subgraph to be infused into data analysis*

```python
def summary_graph(domain, from_database):
    
    domain = str(domain)
    from_database = str(from_database)
    
    # First, let's connect to the right KG database, and locate the domain knowledge inside it
    
    driver = neo4j.GraphDatabase.driver(uri="bolt://localhost:11003", auth=("neo4j","GoPeaks"))
    session = driver.session(database=from_database)
    
    # Next, let's retrieve all the nodes and links from this domain, and generate a results table.

    results_tab = pd.DataFrame({'c1.name': [],
                                'c2.name': [],
                                'correlation': [],
                                'p-value': []
                               })

    path = """MATCH p_c12 = (c1:construct)-[c12:causes]->(c2:construct)
    MATCH p_m1 = (v1:variable)-[m1:measures]->(c1:construct)
    MATCH p_m2 = (v2:variable)-[m2:measures]->(c2:construct)
    MATCH p_r12 = (v1:variable)-[r12:correlates_with]->(v2:variable)
    """

    results = "c1.name, c2.name, r12.correlation, r12.sample_size"

    query = f"""
    {path}
    WHERE c12.paper_id='{domain}'
    RETURN {results}
    """

    results_tab_add = pd.DataFrame(session.run(query)).drop_duplicates().reset_index(drop=True)
    results_tab_add.columns = ['c1.name', 'c2.name', 'correlation', 'sample_size']
    for i, j in zip(results_tab_add['c1.name'], results_tab_add['c2.name']):
        n = int(results_tab_add.loc[(results_tab_add['c1.name']==i) & (results_tab_add['c2.name']==j), 'sample_size'])
        dist = scipy.stats.beta(n/2 - 1, n/2 - 1, loc=-1, scale=2)
        r = float(results_tab_add.loc[(results_tab_add['c1.name']==i) & (results_tab_add['c2.name']==j), 'correlation'])
        results_tab_add.loc[(results_tab_add['c1.name']==i) & (results_tab_add['c2.name']==j), 
                            'p-value'] = 2*dist.cdf(-abs(r))

    results_tab = pd.concat([results_tab,results_tab_add], ignore_index=True)

    # Finally, we are ready to draw the causal KG into a causal graph, with meta-analysis evidence
    
    nodes1 = results_tab['c1.name'].drop_duplicates()
    nodes2 = results_tab['c2.name'].drop_duplicates()
    nodes = pd.concat([nodes1, nodes1], ignore_index=True).drop_duplicates()

    G = gv.Digraph('summary', format='png')

    for i, j in zip(results_tab['c1.name'], results_tab['c2.name']):
        w = float(results_tab.loc[(results_tab['c1.name']==i) & (results_tab['c2.name']==j), 'correlation'])
        p = float(results_tab.loc[(results_tab['c1.name']==i) & (results_tab['c2.name']==j), 'p-value'])
        w = round(w, 3)
        p = round(p, 2)
        if w == 0:
            continue
        elif w < 0:
            G.edge(i, j, **{'penwidth':str(abs(w)*10), 'color': 'red', 'label': f"{w} ({p})"})
        elif w > 0:
            G.edge(i, j, **{'penwidth':str(w*10), 'color': 'green', 'label': f"{w} ({p})"})

    G.render()
    
    note ="""
    Notes: 
    [1] link label shows the mean effect size (p-value in the parenthesis); 
    [2] link arrow shows the direction of causality; 
    [3] link color suggests the sign of the effect size: green for positive, and red for negative;
    [4] link thickness suggests the magnitude of the effect size.
    """
    
    from IPython.display import display, Image
    display(Image(filename='Summary.gv.png'))
    print(note)
```

Let's use the package [DoubleML](https://docs.doubleml.org/stable/index.html) to estimate the average treatment effect of 401(k) eligibility, i.e. `e401`, and participation, i.e. `p401`, on net financial assets `net_tfa`.

## Estimating the Average Treatment Effect of 401(k) Eligibility on Net Financial Assets

Let us first look at the treatment effect of `e401` on net total financial assets. Here are the estimates of the ATE in the linear model

\begin{equation*}
Y = D \alpha + f(X)'\beta+ \epsilon,
\end{equation*}
where $f(X)$ is a dictonary applied to the raw regressors. $X$ contains variables on marital status, two-earner status, defined benefit pension status, IRA participation, home ownership, family size, education, age, and income. 

In the following, let us consider two different models, 

* a basic model specification that includes the raw regressors, i.e., $f(X) = X$, and 

* a flexible model specification, where $f(X)$ includes the raw regressors $X$ and the orthogonal polynomials of degree 2 for the variables family size education, age, and income. 

Let us use the basic model specification whenever we use nonlinear methods, for example regression trees or random forests, and use the flexible model for linear methods such as the lasso. There are, of course, multiple ways how the model can be specified even more flexibly, for example including interactions of variable and higher order interaction. However, for the sake of simplicity we stick to the specification above. Users who are interested in varying the model can adapt the code below accordingly, for example to implement the orignal specification in [Chernozhukov et al. (2018)](https://arxiv.org/abs/1608.00060). 

In the first step, let us report estimates of the average treatment effect (ATE) of 401(k) eligibility on net financial assets both in the partially linear regression (PLR) model and in the interactive regression model (IRM) allowing for heterogeneous treatment effects.

Source: https://docs.doubleml.org/

### The Data Backend: `DoubleMLData`

To start the analysis, we need to initialize the data backend, i.e., a new instance of a [DoubleMLData](https://docs.doubleml.org/dev/api/generated/doubleml.DoubleMLData.html#doubleml.DoubleMLData) object. Now let us implement the regression model by using scikit-learn's `PolynomialFeatures` class.

To implement both models (basic and flexible), here is how to generate two data backends: `data_dml_base` and `data_dml_flex`.

```python
# Set up basic model: Specify variables for data-backend
features_base = ['age', 'inc', 'educ', 'fsize', 'marr',
                 'twoearn', 'db', 'pira', 'hown']

# Initialize DoubleMLData (data-backend of DoubleML)
data_dml_base = dml.DoubleMLData(data,
                                 y_col='net_tfa',
                                 d_cols='e401',
                                 x_cols=features_base)
```

```python
print(data_dml_base)
```

```python
# Set up a model according to regression formula with polynomials
features = data.copy()[['marr', 'twoearn', 'db', 'pira', 'hown']]

poly_dict = {'age': 2,
             'inc': 2,
             'educ': 2,
             'fsize': 2}
for key, degree in poly_dict.items():
    poly = PolynomialFeatures(degree, include_bias=False)
    data_transf = poly.fit_transform(data[[key]])
    x_cols = poly.get_feature_names([key])
    data_transf = pd.DataFrame(data_transf, columns=x_cols)
    
    features = pd.concat((features, data_transf),
                          axis=1, sort=False)

model_data = pd.concat((data.copy()[['net_tfa', 'e401']], features.copy()),
                        axis=1, sort=False)

# Initialize DoubleMLData (data-backend of DoubleML)
data_dml_flex = dml.DoubleMLData(model_data, y_col='net_tfa', d_cols='e401')
```

```python
print(data_dml_flex)
```

```python
y = data['net_tfa']
D = data['e401']
Z = features
```

# Double/Debiased Machine Learning (DML) Methods

### Partially Linear Regression Model (PLR)

Let us start using lasso to estimate the function $g_0$ and $m_0$ in the following PLR model:

\begin{eqnarray}
& Y = D\theta_0 + g_0(X) + \zeta, &\quad E[\zeta \mid D,X]= 0,\\
& D = m_0(X) +  V, &\quad E[V \mid X] = 0.
\end{eqnarray}

To estimate the causal parameter $\theta_0$ here, let us use double machine learning with 3-fold cross-fitting. 
 
Estimation of the nuisance components $g_0$ and $m_0$, is based on the lasso with cross-validated choice of the penalty term , $\lambda$, as provided by [scikit-learn](https://scikit-learn.org). Let us load the learner by initializing instances from the classes [LassoCV](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html) and [LogisticRegressionCV](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegressionCV.html). Hyperparameters and options can be set during instantiation of the learner. The lasso should use that value of $\lambda$ that minimizes the cross-validated mean squared error which is based on 5-fold cross validation.

Now let us start by estimation the ATE in the basic model and then repeat the estimation in the flexible model.

```python
# Initialize learners
Cs = 0.0001*np.logspace(0, 4, 10)
lasso = make_pipeline(StandardScaler(), LassoCV(cv=5, max_iter=10000))
lasso_class = make_pipeline(StandardScaler(),
                            LogisticRegressionCV(cv=5, penalty='l1', solver='liblinear',
                                                 Cs = Cs, max_iter=1000))

np.random.seed(123)
# Initialize DoubleMLPLR model
dml_plr_lasso = dml.DoubleMLPLR(data_dml_base,
                                ml_g = lasso,
                                ml_m = lasso_class,
                                n_folds = 3)

dml_plr_lasso.fit(store_predictions=True)

display(dml_plr_lasso.summary)

# Plot feature importance in two ML models

fig, ax = plt.subplots(1, 2, figsize=(20,5))
fig.suptitle('Feature importance in debiasing MLs', size=20)

ml_m = lasso.fit(data[features_base], y)
ml_m_importance = permutation_importance(ml_m, data[features_base], y, n_repeats=10, random_state=42, n_jobs=2)
ml_m_importance = pd.Series(ml_m_importance.importances_mean, index=features_base).sort_values(ascending=False)

ax[0].bar(ml_m_importance.index.tolist(), ml_m_importance, color=colors)
ax[0].set_title("ML(1): net_tfa = g(features_base)", size=15)
ax[0].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[0].set_xticklabels(ml_m_importance.index.tolist(), fontsize=14, rotation=90)

ml_g = lasso_class.fit(data[features_base], D)
ml_g_importance = permutation_importance(ml_g, data[features_base], D, n_repeats=10, random_state=42, n_jobs=2)
ml_g_importance = pd.Series(ml_g_importance.importances_mean, index=features_base).sort_values(ascending=False)

ax[1].bar(ml_g_importance.index.tolist(), ml_g_importance, color=colors)
ax[1].set_title("ML(2): e401 = m(features_base)", size=15)
ax[1].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[1].set_xticklabels(ml_g_importance.index.tolist(), fontsize=14, rotation=90)

plt.show()
plt.close()
```

```python
# Estimate the ATE in the flexible model with lasso
np.random.seed(123)

# Get the scores in the 'two' ML steps
g_plr_lasso_score = lasso.fit(Z,y).score(Z,y)
m_plr_lasso_score = lasso_class.fit(Z,D).score(Z,D)

dml_plr_lasso = dml.DoubleMLPLR(data_dml_flex,
                                ml_g = lasso,
                                ml_m = lasso_class,
                                n_folds = 3)

dml_plr_lasso.fit(store_predictions=True)
lasso_summary = dml_plr_lasso.summary

display(lasso_summary)

# Plot feature importance in two ML models

fig, ax = plt.subplots(1, 2, figsize=(20,5))
fig.suptitle('Feature importance in debiasing MLs', size=20)

ml_m = lasso.fit(Z, y)
ml_m_importance = permutation_importance(ml_m, Z, y, n_repeats=10, random_state=42, n_jobs=2)
ml_m_importance = pd.Series(ml_m_importance.importances_mean, index=Z.columns).sort_values(ascending=False)

ax[0].bar(ml_m_importance.index.tolist(), ml_m_importance, color=colors)
ax[0].set_title("ML(1): net_tfa = g(features)", size=15)
ax[0].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[0].set_xticklabels(ml_m_importance.index.tolist(), fontsize=14, rotation=90)

ml_g = lasso_class.fit(Z, D)
ml_g_importance = permutation_importance(ml_g, Z, D, n_repeats=10, random_state=42, n_jobs=2)
ml_g_importance = pd.Series(ml_g_importance.importances_mean, index=Z.columns).sort_values(ascending=False)

ax[1].bar(ml_g_importance.index.tolist(), ml_g_importance, color=colors)
ax[1].set_title("ML(2): e401 = m(features)", size=15)
ax[1].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[1].set_xticklabels(ml_g_importance.index.tolist(), fontsize=14, rotation=90)

plt.show()
plt.close()
```

Alternatively, we can repeat this procedure with other machine learning methods, see, e.g., [scikit-learn](https://scikit-learn.org).

### Interactive Regression Model (IRM)

Next, let us consider estimation of average treatment effects when treatment effects are fully heterogeneous:

\begin{eqnarray}
& Y = g_0(D,X) + U, &\quad E[U\mid X,D] = 0,\\
& D = m_0(X) + V, &\quad E[V\mid X] = 0.
\end{eqnarray}

To reduce the disproportionate impact of extreme propensity score weights in the interactive model,
let us trim the propensity scores which are close to the bounds.

## Local Average Treatment Effects of 401(k) Participation on Net Financial Assets

### Interactive IV Model (IIVM)

The examples above estimated the average treatment effect of *eligibility* on financial asset holdings. Now, let us consider estimation of local average treatment effects (LATE) of *participation* using eligibility as an instrument for the participation decision. Under appropriate assumptions, the LATE identifies the treatment effect for so-called compliers, i.e., individuals who would only participate if eligible and otherwise not participate in the program. 

As before, $Y$ denotes the outcome `net_tfa`, and $X$ is the vector of covariates. Here `e401` is a binary instrument for the treatment variable `p401`. Here the structural equation model is:

\begin{eqnarray}
& Y = g_0(Z,X) + U, &\quad E[U\mid Z,X] = 0,\\
& D = r_0(Z,X) + V, &\quad E[V\mid Z, X] = 0,\\
& Z = m_0(X) + \zeta, &\quad E[\zeta \mid X] = 0.
\end{eqnarray}

```python
# Initialize DoubleMLData with an instrument

# Basic model
data_dml_base_iv = dml.DoubleMLData(data,
                                    y_col='net_tfa',
                                    d_cols='p401',
                                    z_cols='e401',
                                    x_cols=features_base)

print(data_dml_base_iv)
```

```python
# Flexible model
model_data = pd.concat((data.copy()[['net_tfa', 'e401', 'p401']], features.copy()),
                        axis=1, sort=False)

data_dml_iv_flex = dml.DoubleMLData(model_data,
                                    y_col='net_tfa',
                                    d_cols='p401',
                                    z_cols='e401')

print(data_dml_iv_flex)
```

```python
# Lasso
lasso = make_pipeline(StandardScaler(), LassoCV(cv=5, max_iter=20000))

# Initialize DoubleMLIRM model
np.random.seed(123)
dml_iivm_lasso = dml.DoubleMLIIVM(data_dml_iv_flex,
                                  ml_g = lasso,
                                  ml_m = lasso_class,
                                  ml_r = lasso_class,
                                  subgroups = {'always_takers': False,
                                             'never_takers': True},
                                  trimming_threshold = 0.01,
                                  n_folds = 3)
dml_iivm_lasso.fit(store_predictions=True)
lasso_summary = dml_iivm_lasso.summary

display(lasso_summary)
```

## Example 1: Remaining confounding bias due to missing key confounders

```python
# Draw a Causal Bayesian DAG
import graphviz

g = graphviz.Digraph('incorrect_1', format='png')

# orange for main relationship to be estimated
g.attr('node', style='filled', color='orange', shape='oval')
g.attr('edge', color='orange')
g.edge('e401', 'net_tfa')

# Red for confounders
g.attr('node', style='filled', color='lightblue', shape='oval')
g.attr('edge', color='lightblue')
for i in ['age','educ','fsize','marr','twoearn','hown']:
    g.node(i)
    g.edge(i, 'e401')
    
g.attr('edge', color='lightblue')
for i in ['age','educ','fsize','marr','twoearn','hown']:
    g.node(i)
    g.edge(i, 'net_tfa')
    
# Yellow for missing confounders
g.attr('node', style='filled', color='red', shape='oval')
for i in ['inc','db','pira']:
    g.node(i)
    
g.render()

from IPython.display import display, Image
display(Image(filename='incorrect_1.gv.png'))
```

```python
# Set up basic model: Specify variables for data-backend
features_base = ['age','inc','educ','fsize','marr','twoearn','db','pira','hown']

features_base_x = ['age', 'educ', 'fsize', 'marr','twoearn', 'hown']

# Initialize DoubleMLData (data-backend of DoubleML)
data_dml_base = dml.DoubleMLData(data,
                                 y_col='net_tfa',
                                 d_cols='e401',
                                 x_cols=features_base)

data_dml_base_x = dml.DoubleMLData(data,
                                 y_col='net_tfa',
                                 d_cols='e401',
                                 x_cols=features_base_x)

y = data['net_tfa']
D = data['e401']
Zx = data[features_base_x]

# Random Forest
randomForest = RandomForestRegressor(
    n_estimators=500, max_depth=7, max_features=3, min_samples_leaf=3)
randomForest_class = RandomForestClassifier(
    n_estimators=500, max_depth=5, max_features=4, min_samples_leaf=7)

np.random.seed(123)

# Get the scores in the 'two' ML steps
g_plr_forest_score = randomForest.fit(Zx,y).score(Zx,y)
m_plr_forest_score = randomForest_class.fit(Zx,D).score(Zx,D)

dml_plr_forest = dml.DoubleMLPLR(data_dml_base,
                                 ml_g = randomForest,
                                 ml_m = randomForest_class,
                                 n_folds = 3)
dml_plr_forest.fit(store_predictions=True)
forest_summary = dml_plr_forest.summary

dml_plr_forest_x = dml.DoubleMLPLR(data_dml_base_x,
                                 ml_g = randomForest,
                                 ml_m = randomForest_class,
                                 n_folds = 3)
dml_plr_forest_x.fit(store_predictions=True)
forest_summary_x = dml_plr_forest_x.summary

plr_summary = pd.concat((forest_summary, forest_summary_x))
plr_summary.index = ['debiased forest', 'biased forest']
display(plr_summary)

diff = round(float(forest_summary_x['coef'] - forest_summary['coef']),2)
if diff > 0:
    print('The average causal effect is over-estimated by ${}.'.format(abs(diff)))
else:
    print('The average causal effect is under-estimated by ${}.'.format(abs(diff)))

print('\nSummary: ML Scores in Debiasing the Measures')
print('R2(net_tfa = g(features)):', g_plr_forest_score)
print('F1(e401 = m(features)):', m_plr_forest_score)

# Plot feature importance in two ML models

fig, ax = plt.subplots(1, 2, figsize=(20,5))
fig.suptitle('Feature importance in debiasing MLs', size=20)

ml_m = randomForest.fit(Zx, y)
ml_m_importance = permutation_importance(ml_m, Zx, y, n_repeats=10, random_state=42, n_jobs=2)
ml_m_importance = pd.Series(ml_m_importance.importances_mean, index=Zx.columns).sort_values(ascending=False)

ax[0].bar(ml_m_importance.index.tolist(), ml_m_importance, color=colors)
ax[0].set_title("ML(1): net_tfa = g(features)", size=15)
ax[0].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[0].set_xticklabels(ml_m_importance.index.tolist(), fontsize=14, rotation=90)

ml_g = randomForest_class.fit(Zx, D)
ml_g_importance = permutation_importance(ml_g, Zx, D, n_repeats=10, random_state=42, n_jobs=2)
ml_g_importance = pd.Series(ml_g_importance.importances_mean, index=Zx.columns).sort_values(ascending=False)

ax[1].bar(ml_g_importance.index.tolist(), ml_g_importance, color=colors)
ax[1].set_title("ML(2): e401 = m(features)", size=15)
ax[1].set_ylabel("Mean accuracy decrease", fontsize=14)
ax[1].set_xticklabels(ml_g_importance.index.tolist(), fontsize=14, rotation=90)

plt.show()
plt.close()
```

## Example 2: Introducing collider bias due to including wrong confounders

```python
# Draw a Causal Bayesian DAG
import graphviz

g = graphviz.Digraph('incorrect_2', format='png')

# orange for main relationship to be estimated
g.attr('node', style='filled', color='orange', shape='oval')
g.attr('edge', color='orange')
g.edge('e401', 'net_tfa')

# Wrong confounder
g.attr('node', style='filled', color='red', shape='oval')
g.attr('edge', color='red')
g.edge('tw', 'e401')
g.edge('tw', 'net_tfa')

# Red for confounders
g.attr('node', style='filled', color='lightblue', shape='oval')
g.attr('edge', color='lightblue')
for i in ['age','educ','fsize','marr','twoearn','hown', 'inc','db','pira']:
    g.node(i)
    g.edge(i, 'e401')
    
g.attr('edge', color='lightblue')
for i in ['age','educ','fsize','marr','twoearn','hown', 'inc','db','pira']:
    g.node(i)
    g.edge(i, 'net_tfa')
    
    
g.render()

from IPython.display import display, Image
display(Image(filename='incorrect_2.gv.png'))
```
